Problem Statement
Environmental justice (EJ) seeks to ensure that all communities receive fair treatment and meaningful involvement in environmental decision‑making, regardless of socioeconomic or demographic characteristics. In Michigan, the Department of Environment, Great Lakes, and Energy (EGLE) has developed Environmental Justice screening tools that combine measures of income, race, health, and pollution exposure to identify areas of potential concern. While these tools are effective at highlighting broad regional disparities, they are primarily interpreted at larger geographic scales, which can mask important patterns of environmental burden within cities.
Wayne County and particularly the City of Detroit have a long history of industrial development and contains numerous sources of environmental pollution, including hazardous sites and air pollution sources. As a result, residents in certain neighborhoods may face disproportionately high environmental risks. However, without fine‑scale analysis, it is difficult to determine whether these environmental burdens are evenly distributed across the city or concentrated within specific census tracts.
The problem addressed in this project is the lack of small‑scale, census‑tract‑level analysis that integrates multiple environmental exposure and ecological burden indicators with population data. This project seeks to fill that gap by examining environmental burden patterns across Detroit census tracts, identifying areas of elevated cumulative risk, and estimating how many residents live within those high‑burden areas. By applying spatial analysis and percentile‑based classification methods, the study aims to reveal localized environmental justice concerns that may not be evident in broader statewide assessments.

Research Objectives
The primary objective of this research is to evaluate environmental justice conditions in Detroit, Michigan by examining how environmental exposures and ecological burdens are distributed across census tracts. The study adopts a small‑scale, census‑tract‑level approach to capture localized variations in environmental risk that may be obscured in regional or statewide assessments.
Specifically, this project aims to:


Assess the spatial distribution of environmental exposure and ecological burden across Detroit census tracts using multiple indicators, including air pollution (PM2.5, NO₂, and O₃) and the presence of solid waste facilities, contaminated sites, and cleanup sites.


Develop composite indices of environmental burden by integrating air exposure and ecological burden indicators, providing a cumulative measure of environmental risk rather than relying on single variables.


Identify census tracts with moderate, high, and very high environmental burden using percentile‑based classification thresholds (25th, 50th, and 75th percentiles).


Examine the spatial concentration of high‑burden census tracts to determine whether environmental risks are evenly distributed or clustered in specific parts of the city.


Estimate the population affected by elevated environmental burden by linking population data to identified high‑burden census tracts and calculating the number of residents living in these areas.


Through these objectives, the study seeks to provide a spatially explicit understanding of environmental burden patterns in Detroit and contribute empirical evidence to ongoing discussions of environmental justice at the neighborhood scale.

```python
def plot_choropleth(
    gdf,
    column,
    title,
    cmap="YlOrBr",
    scheme="quantiles",
    k=5,
    legend_loc="lower right"
):
    """
    Plot a choropleth map for a GeoDataFrame.

    Parameters
    ----------
    gdf : GeoDataFrame
        GeoDataFrame containing geometry and data.
    column : str
        Column to visualize.
    title : str
        Map title.
    cmap : str, optional
        Matplotlib colormap (default is 'YlOrBr').
    scheme : str, optional
        Classification scheme ('quantiles', 'naturalbreaks', etc.).
    k : int, optional
        Number of classes.
    legend_loc : str, optional
        Legend location.
    """

    fig, ax = plt.subplots(figsize=(10, 8))

    gdf.plot(
        column=column,
        scheme=scheme,
        k=k,
        cmap=cmap,
        linewidth=0.4,
        edgecolor="black",
        legend=True,
        ax=ax,
        legend_kwds={"loc": legend_loc}
    )

    ax.set_title(title, fontsize=14, weight="bold")
    ax.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
plot_choropleth(
    detroit_tracts_pop,
    "Total Population",
    "Spatial Distribution of Population in Detroit"
)
```

This choropleth map illustrates the spatial distribution of population across census tracts in Detroit. Each tract is shaded based on its total population, with lighter colors representing lower populations and darker brown tones indicating higher concentrations. The map shows that population is unevenly distributed, with clusters of higher population in certain areas and lower population in others, reflecting patterns of residential density, land use, and possible vacancy or industrial zones.

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Common settings
column = "Total Population"
cmap = "YlOrBr"
k = 5

# ---- Quantile Map (NO legend) ----
detroit_tracts_pop.plot(
    column=column,
    scheme="quantiles",
    k=k,
    cmap=cmap,
    linewidth=0.4,
    edgecolor="black",
    legend=False,          
    ax=axes[0]
)

axes[0].set_title("Quantile Classification", fontsize=13)
axes[0].axis("off")

# ---- Natural Breaks Map (NO legend) ----
detroit_tracts_pop.plot(
    column=column,
    scheme="naturalbreaks",
    k=k,
    cmap=cmap,
    linewidth=0.4,
    edgecolor="black",
    legend=False,        
    ax=axes[1]
)

axes[1].set_title("Natural Breaks (Jenks)", fontsize=13)
axes[1].axis("off")

# ---- Create ONE shared legend ----
norm = mpl.colors.Normalize(
    vmin=detroit_tracts_pop[column].min(),
    vmax=detroit_tracts_pop[column].max()
)

sm = mpl.cm.ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])

cbar = fig.colorbar(
    sm,
    ax=axes,
    orientation="vertical",
    fraction=0.03,
    pad=0.04
)

cbar.set_label("Total Population", fontsize=11)

# ---- Overall title ----
fig.suptitle(
    "Spatial Distribution of Population by Census Tract, Detroit",
    fontsize=15
)

plt.show()
```

The maps compare two classification methods for visualizing population by census tract in Detroit. The quantile classification (left) assigns an equal number of tracts to each class, emphasizing relative differences across the city. The natural breaks (Jenks) method (right) groups tracts based on inherent gaps in the data, highlighting areas with distinctly higher or lower populations. While both show uneven population distribution, natural breaks better emphasize clusters of high‑population tracts, whereas quantiles provide a more balanced comparison across all tracts.

```python
def plot_burden_threshold(gdf, column, title):
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plot all tracts (background)
    gdf.plot(
        ax=ax,
        color="lightgray",
        linewidth=0.2,
        edgecolor="white"
    )
    
    # Plot high-burden tracts
    gdf[gdf[column]].plot(
        ax=ax,
        color="#d73027",  # strong red
        linewidth=0.4,
        edgecolor="black"
    )
    
    ax.set_title(title, fontsize=14, weight="bold")
    ax.axis("off")
    plt.tight_layout()
    plt.show()
```

CONCLUSION
This project examined environmental justice conditions in Detroit by analyzing environmental exposure, ecological burdens, and population distribution at the census‑tract level. By integrating census population data, satellite‑derived air pollution measures (NO₂ and O₃), and multiple ecological burden indicators, the analysis revealed that environmental risks are unevenly distributed across the city. Composite environmental burden scores showed clear spatial clustering, with certain neighborhoods experiencing significantly higher cumulative exposure to environmental hazards than others.
The percentile‑based analysis further demonstrated that as thresholds increased from the 25th to the 75th percentile, high‑burden census tracts became increasingly concentrated in specific areas rather than being evenly spread citywide. This pattern indicates that environmental burdens in Detroit are not random, but instead reflect localized concentrations of risk, consistent with key environmental justice concerns.


Key Insights
Several important insights emerged from this analysis. First, while air pollution indicators such as NO₂ and O₃ showed relatively broad regional patterns, ecological burden indicators (solid waste sites, contaminated sites, and cleanup sites) were far more unevenly distributed. This suggests that localized land‑use history and industrial activity play a critical role in shaping environmental inequities.
Second, the cumulative environmental burden index proved effective in capturing multi‑dimensional risk. Combining multiple pollutants and ecological stressors into a single score highlighted neighborhoods that may not appear extreme when considering any single variable alone but face elevated cumulative exposure. This reinforces the importance of using composite indices rather than single indicators in environmental justice research.


One surprising finding was the extent to which moderate environmental burden was widespread across the city. At the 25th percentile threshold, a large proportion of census tracts were classified as having elevated environmental burden, suggesting that environmental risk is a common experience for many Detroit residents, not just those in the highest‑risk neighborhoods.
Another notable observation was the sharp reduction in the number of tracts—and associated population—at the 75th percentile threshold. This indicates that while severe environmental burden affects a smaller portion of census tracts, it is highly concentrated, raising concerns about disproportionate impacts on residents living in those areas.

Limitations
This study has several limitations. The analysis relied on census‑tract‑level aggregation, which may mask within‑tract variation in exposure.
Satellite‑derived air pollution data reflect average conditions over a short time period and may not capture long‑term exposure variability. Finally, income and health outcome data were not explicitly modeled, limiting the ability to directly link environmental burden to social or health disparities.
